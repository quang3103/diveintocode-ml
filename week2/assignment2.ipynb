```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
csv_path = "mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
print("File content:\n{}".format(fuji[130: 140]))
```

```python
#problem 1
plt.title("Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation (m)")
plt.plot(fuji[:, 0], fuji[:, 3])
plt.show()
plt.clf()
```

```python
#problem 2
x = fuji[:, 0] #point number
y = fuji[:, 3] #elevation
def compute_slope_of_a_point(pointNumber = 1):
    d = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    return d[pointNumber - 1]
print(compute_slope_of_a_point(136))
```

```python
#problem 3
def compute_destination(alpha, currentLocation):
    hyperParameter = alpha
    gradient = compute_slope_of_a_point(currentLocation)
    return round(currentLocation - hyperParameter * gradient)
print(compute_destination(0.2, 136))
```

```python
#problem 4
locationList = [136]
def go_down_mountain(alpha = 0.2, position = 136):
    global locationList
    locationList = [position]
    while True:
        locationList.append(compute_destination(alpha, locationList[-1]))
        if (y[locationList[-1]] == y[locationList[-2]]):
            break
go_down_mountain()
print(locationList)
```

```python
#problem 5
plt.title("Going down Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation (m)")
plt.plot(fuji[:, 0], fuji[:, 3])
plt.plot(locationList, [y[locationList[i]] for i in range(len(locationList))], color = 'red')
plt.show()
plt.clf()
```

```python
#problem 6
go_down_mountain(0.2, 125)
plt.title("Going down Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation (m)")
plt.plot(fuji[:, 0], fuji[:, 3])
plt.plot(locationList, [y[locationList[i]] for i in range(len(locationList))], color = 'red')
plt.show()
plt.clf()
```
