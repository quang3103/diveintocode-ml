```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

# problem 1

```python
data = load_iris()
x = pd.DataFrame(data.data, columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print(x)
y = pd.DataFrame(data.target, columns = ['Species'])
print(data.target_names)
print(y)
```

# problem 2

```python
df = pd.concat([x, y], axis=1)
print(df)
```

# problem 3

```python
print("The 4th sample from the beginning:")
print(df.head(4).info())
print("Total number of samples for each label:")
print(df.Species.value_counts())
print("Check if there is a missing value in the feature quantity:")
print(df.isnull().sum())
print("Display the mean, standard deviation, and quartiles of the feature values at once")
print(df.describe())
```

# problem 4


The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. Two of the three species were collected in the Gasp√© Peninsula "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus".

# problem 5

The difference between .loc and .iloc is that .loc is label-based, while .iloc is integer index based (located by sequential position). Therefore, you can use .loc when you are working labels, and .iloc when you find some elements with integer index.

```python
print("+ First Way:")
print(df['sepal_width'])
print("+ Second Way:")
print(df.iloc[:, 1])
print("+ Third Way:")
print(df.loc[:, 'sepal_width'])
print("Data from the 50th to 99th row")
print(df.iloc[50:99])
print("Data from the 50th to 99th data of the petal_length column")
print(df.iloc[50:99, 2])
print("Data with a petal_width value of 0.2")
print(df.petal_width == 0.2)
```
