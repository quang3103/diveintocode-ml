```python
#problem 1
import matplotlib.pyplot as plt
numberOfGrains = [1]
totalNumberOfGrains = [1]
for i in range(100):
    numberOfGrains.append(numberOfGrains[-1] * 2)
    totalNumberOfGrains.append(numberOfGrains[-1] + totalNumberOfGrains[-1])
#print(totalNumberOfGrains)
plt.xlabel("Number of Days")
plt.ylabel("Number of Grains")
plt.plot([i for i in range(1,101)], numberOfGrains[1:], color = 'blue', marker = '*', mec = 'yellow')
plt.plot([i for i in range(1,101)], totalNumberOfGrains[1:], color = 'green', marker = '^', mec = 'red')
plt.show()
```

```python
#problem 2
def compute_sorori_shinzaemon(day = 100):
    numberOfGrains = [1]
    totalNumberOfGrains = [1]
    for i in range(day):
        numberOfGrains.append(numberOfGrains[-1] * 2)
        totalNumberOfGrains.append(numberOfGrains[-1] + totalNumberOfGrains[-1])
    return numberOfGrains[1:], totalNumberOfGrains[1:]

list_n_grains, list_total_grains = compute_sorori_shinzaemon(10)
print(list_n_grains)
print(list_total_grains)
plt.xlabel("Number of Days")
plt.ylabel("Number of grains")
plt.plot([i for i in range(1,len(list_n_grains)+1)], list_n_grains, color = 'blue', marker = '*', mec = 'purple')
plt.plot([i for i in range(1,len(list_total_grains)+1)], list_total_grains, color = 'green', marker = '^', mec = 'red')
plt.show()
```
