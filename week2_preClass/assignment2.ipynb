```python
import random
import numpy as np
import matplotlib.pyplot as plt

#problem 1
np.random.seed(0)
mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]]
data = np.random.multivariate_normal(mean = mean, cov = cov, size = 500)
print(data.shape)
print(data)
```

```python
#problem 2

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter")
plt.scatter([data[i][0] for i in range(len(data))], [data[i][1] for i in range(len(data))])
plt.show()
```

```python
#problem 3

plt.xlabel("X")
plt.ylabel("Frequency")
plt.title("Histogram of X")
plt.xlim(-5, 3)
plt.hist([data[i][0] for i in range(len(data))], bins = 50)
plt.show()

plt.xlabel("Y")
plt.ylabel("Frequency")
plt.title("Histogram of Y")
plt.xlim(-5, 3)
plt.hist([data[i][1] for i in range(len(data))], bins = 50)
plt.show()
```

```python
#problem 4

mean1 = [0, -3]
data1 = np.random.multivariate_normal(mean = mean1, cov = cov, size = 500)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter")
plt.scatter([data[i][0] for i in range(len(data))], [data[i][1] for i in range(len(data))], label = "1")
plt.scatter([data1[i][0] for i in range(len(data1))], [data1[i][1] for i in range(len(data1))], label = "2")
plt.legend(loc="upper right")
plt.show()
```

```python
#problem 5

array = np.concatenate((data, data1), axis = 0)
print(array.shape)
print(array)
```

```python
#problem 6

label = np.concatenate((np.zeros((500, 1)), np.ones((500, 1))), axis = 0)
array1 = np.concatenate((array, label), axis = 1)
print(array1.shape)
print(array1)
```
